```python
import base64
img_1 = open("couple.jpeg", 'rb')
base_encoded = base64.b64encode(img_1.read()).decode()
```

```python
base_encoded = 'data:image/jpeg;base64,' + base_encoded
base_encoded
```

```python
import requests
url = 'http://0.0.0.0:8080/detect_faces'
payload = {'source_face_b64': base_encoded}
temp = requests.post(url, files=payload)
```
